```python
# Exercise 1
def func1(a, b):
  result = 0
  for i in range(0, b):
    result += a
  return result
```

```python
# Exercise 2
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result
```

```python
# Sample function call: func2(2, 4)
print(func2(2, 4))
# this is an iterative function which used to give the power of the number
import time
start=time.time()
func2(2, 4)
end=time.time()
print("running time",(end-start)*1000)
```

```python
# Exercise 3
def func3(num):
   result = 0
   while num > 0:
    result = result + num % 10
    num //= 10

   return result
```

```python
# Sample function call: func3(1234)
print(func3(1234))
#it performs the addition of the digits of the given number
import time
start=time.time()
func3(1234)
end=time.time()
print("running time",(end-start)*1000)
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
print(func4([1, 2, 3, 4], [2, 4, 6, 8]))
#the function use to give the no of times duplicate element occur in the list
import time
start=time.time()
func4([1, 2, 3, 4], [2, 4, 6, 8])
end=time.time()
print("running time:",(end-start)*1000)
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
        return None
```

```python
# Sample function call: func5(16)
print(func5(16))
import time
start=time.time()
func5(16)
end=time.time()
print("running time:",(end-start)*1000)
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
start=time.time()
x=print(func8([3,1,2,7,8,6]))
end=time.time()
print("running time:",end-start)
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
start=time.time()
x=print(func12([1, 2, 3, 4, 5], 2))
end=time.time()
print("running time:",end-start)
```

Programs on 1-d arrays like - sum of elements of array, searching an element inarray, finding minimum and maximum element in array, count the number ofeven and odd numbers in array. For all such programs, also find the timeComplexity, compare if there are multiple methods

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  total = 0
  for num in arr:
    total += num
    return total
```

```python
#call the function
#arr = [12, 3, 4, 15,45,78,450]
#print(sum_arr)
arr = [12,3,4,15,45,78,450]
result = sum_arr(arr)
print(result)
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
  if len(arr) == 0:
    return None, None
  maximum = arr[0]
  minimum = arr[0]
  for i in range(1, len(arr)):
    if arr[i] >maximum:
      maximum = arr[i]
    elif arr[i] < minimum:
      minimum = arr[i]
  return maximum, minimum
arr = [12, 3, 4, 15, 45, 78, 450]
max_val, min_val = max_min(arr)
print("Maximum element:", max_val)
print("Minimum element:", min_val)
```

```python
# call sum_arr to find sum of elements in given array for arr = [12, 1234, 45, 67, 1]
def sum_arr(arr):
    total = 0
    for num in arr:
        total += num
    return total
arr = [12, 1234, 45, 67, 1]
result = sum_arr(arr)
print("The sum of the array is:", result)
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
    even_count = 0
    odd_count = 0
    for num in arr:
      if num % 2 == 0:
            even_count += 1
      else:
            odd_count += 1
    return even_count, odd_count
arr = [12, 1234, 45, 67, 1]
evens, odds = Counting_EvenOdd(arr)
print(f"Even numbers: {evens}")
print(f"Odd numbers: {odds}")
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
```

```python
# call the function liner_Search for array = [1, 3, 5, 7, 9]
arr = [1, 3, 5, 7, 9]
target = 7
result = linear_Search(arr, target)
target_missing = 4
result_missing = linear_Search(arr, target_missing)
print(f"Index of {target}: {result}")
print(f"Index of {target_missing}: {result_missing}")
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
arr = [1, 3, 5, 7, 9, 14, 15, 16, 45]
target = 14
result = Binary_Search(arr, target)
print(f"Index of {target} in sorted array: {result}")
```
